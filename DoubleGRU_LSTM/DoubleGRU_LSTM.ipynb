```python
X = sequences_augmented
y= to_categorical(labels_augmented).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)    
```

```python
from tensorflow import keras
from keras.layers import Dense, LSTM
from keras.models import Sequential
from keras.callbacks import TensorBoard

log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, GRU, Dense
from keras.optimizers import Adam

# Define the model
model_1 = Sequential()
model_1.add(GRU(64, return_sequences=True, input_shape=(30, 1662)))  # GRU layer
model_1.add(GRU(64, return_sequences=True, input_shape=(30, 1662)))  # GRU layer
model_1.add(LSTM(128))  # LSTM layer
model_1.add(Dense(65, activation='relu'))
model_1.add(Dense(65, activation='softmax'))  # 64 classes

# Compile the model
model_1.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

from keras.callbacks import EarlyStopping

# Define the early stopping monitor
early_stopping_monitor = EarlyStopping(patience=5)

# Fit the model to the data
model_1.fit(X_train, y_train, epochs=150, validation_split=0.2, callbacks=[early_stopping_monitor])
```
