```python
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
# imports

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import time
import mediapipe as mp
```

```python
# Load sequences and labels from disk

sequences = np.load('../sequences.npy')
labels = np.load('../labels.npy')

sequences_augmented = np.load('../sequences_augmented.npy')
labels_augmented = np.load('../labels_augmented.npy')

################!!!!!!!!!!!!!!!!!!!!! they are now nparrays instead of lists, so no need to convert them to nparrays again in the code below
```

```python
X = sequences_augmented
y= to_categorical(labels_augmented).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)    
```

```python
from tensorflow import keras
from keras.layers import Dense, LSTM
from keras.models import Sequential
from keras.callbacks import TensorBoard

log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, GRU, Dense, Dropout
from keras.optimizers import Adam
from keras.regularizers import l2
from keras.callbacks import EarlyStopping

# Define the model
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(30, 1662), dropout=0.5))  # LSTM layer with dropout
model.add(GRU(128, kernel_regularizer=l2(0.01)))  # GRU layer with L2 regularization
model.add(Dense(65, activation='relu', kernel_regularizer=l2(0.01)))  # Dense layer with L2 regularization
model.add(Dense(65, activation='softmax'))  # 64 classes

# Compile the model
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Define the early stopping monitor
early_stopping_monitor = EarlyStopping(patience=10)

# Fit the model to the data
model.fit(X_train, y_train, epochs=150, validation_split=0.2, callbacks=[early_stopping_monitor])
```

```python
model.save('../actions/actions_lstm_gru.h5')  
```

```python
from sklearn.metrics import accuracy_score, classification_report, multilabel_confusion_matrix


y_pred = model.predict(X_test)
ytrue = np.argmax(y_test, axis=1).tolist()
ypred = np.argmax(y_pred, axis=1).tolist()

print(accuracy_score(ytrue, ypred))
multilabel_confusion_matrix(ytrue, ypred)

# Accuracy Score 0.8707692307692307
```
