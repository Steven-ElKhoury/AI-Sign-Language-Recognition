```python
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
# imports

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import time
import mediapipe as mp
```

```python
# Load sequences and labels from disk
sequences = np.load('../sequences.npy')
labels = np.load('../labels.npy')

sequences_augmented = np.load('../sequences_augmented.npy')
labels_augmented = np.load('../labels_augmented.npy')

```

```python
X = sequences_augmented
#output in one hot encoded format
y= to_categorical(labels_augmented).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)    
```

```python
from tensorflow import keras
from keras.layers import Dense, LSTM
from keras.models import Sequential
from keras.callbacks import TensorBoard

log_dir = os.path.join('../Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, LSTM, Dense
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping

# Define the model
model_cnn = Sequential()
model_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(30, 1662)))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(LSTM(128, return_sequences=True))
model_cnn.add(LSTM(128))
model_cnn.add(Dense(65, activation='softmax'))

# Compile the model
optimizer = Adam(learning_rate=0.001)
early_stopping = EarlyStopping(monitor='val_loss', patience=5)
model_cnn.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

model_cnn.fit(X_train, y_train, epochs=100, callbacks=[tb_callback, early_stopping], validation_split=0.2)

```

```python
results_cnn = model_cnn.predict(X_test)
```

```python
model_cnn.save('../actions/actions_cnn.h5')
```

```python
from sklearn.metrics import accuracy_score, classification_report, multilabel_confusion_matrix


y_pred_cnn= model_cnn.predict(X_test)
ytrue = np.argmax(y_test, axis=1).tolist()
ypred_cnn = np.argmax(y_pred_cnn, axis=1).tolist()

print(accuracy_score(ytrue, ypred_cnn))
multilabel_confusion_matrix(ytrue, ypred_cnn)
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, GRU
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping

# Define the model
model_cnn = Sequential()
model_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(30, 1662)))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(GRU(128, return_sequences=True))
model_cnn.add(GRU(128))
model_cnn.add(Dense(65, activation='softmax'))

# Compile the model
optimizer = Adam(learning_rate=0.001)
early_stopping = EarlyStopping(monitor='val_loss', patience=5)
model_cnn.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

model_cnn.fit(X_train, y_train, epochs=100, callbacks=[tb_callback, early_stopping], validation_split=0.2)

```
